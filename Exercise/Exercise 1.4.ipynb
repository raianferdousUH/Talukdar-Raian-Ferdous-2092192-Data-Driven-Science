```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Input= [ [np.random.normal() for i in range(100)]    
    for j in range(100) ]
Input= np.array(Input)

U, S, Vt = np.linalg.svd(Input,full_matrices=0) #Economy SVD

S = np.diag(S)

figure1 = plt.figure()
axis1 = figure1.add_subplot(121)
axis1.semilogy(S,'*',color='r')    
axis2 = figure1.add_subplot(122)
axis2.plot(np.cumsum(S)/np.sum(S),'-o',color='g')
```
