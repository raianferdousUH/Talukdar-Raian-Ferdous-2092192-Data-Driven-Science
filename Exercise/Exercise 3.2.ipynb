```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy.stats import uniform
import math
```

```python
fig, axes = plt.subplots() 


board=plt.Rectangle((-1,-1),2,2,edgecolor='black',facecolor='none',alpha=0.5)
circle = plt.Circle((0, 0), 1, edgecolor='b',facecolor='none',alpha=0.6)
x=np.linspace(-1,1,10000) #10000 points
y=uniform.rvs(-1,2,10000) #uniform distribution
axes = plt.gca() #get current axes

axes.add_patch(board)
axes.add_patch(circle)
axes.scatter(np.linspace(-1,1,10000),uniform.rvs(-1,2,10000),color='y',alpha=0.1)
axes.scatter(x,y,color='y',alpha=0.1)
```

```python
#sqrt = math.sqrt()
rad= x**2+y**2
inc=np.where(rad<1)
inp=rad[inc]
p=len(inp)
fraction=p/10000
print(fraction)
```

```python
data = pd.DataFrame(np.vstack((x, y)).reshape(-1,2), columns=['1','2'])
dim = np.arange(3,11,1)
frac = np.array(fraction)

for i in dim:
    data[i]=uniform.rvs(-1,2,10000)
    fraction=data[((data**2).sum(axis=1)<1)]
    frac=np.append(frac,fraction.shape[0]/10000)

dims=np.arange(2,11,1)
plt.figure(figsize=(10,4))
plt.plot(dims,frac)
plt.title(label='Fraction of dart balls within the circle')
plt.xlabel('Dimension numbers')
plt.ylabel('Dart ball fractions')
```
